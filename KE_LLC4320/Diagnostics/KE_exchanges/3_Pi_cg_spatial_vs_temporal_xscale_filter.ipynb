```python
gridx = xgcm.Grid(ds, coords={'X':{'center':'lon', 'left':'lon_g'},
                              'Y':{'center':'lat', 'left':'lat_g'}}, periodic=['X', 'Y'])
print(gridx)
```

# 2. Calculate $\Pi_\omega$ using the Coarse-graining method (using a spatial filter)

## KE flux $\Pi_\omega$ at 8-km, 30 km and 50-km 

```python
out_dir_store = '/home1/scratch/xyu/KE/Pi_omega/'
VV = ['PI_cg_8km','PI_cg_30km','PI_cg_50km']
Grid_num = [4, 15, 25]
depth_all = range(52)
```

```python
# 22 mins for 8 km
# 48 mins for 20 km 
# 130 mins for 50 km

for vv,grid_num in zip(VV,Grid_num):
    
    for depth in depth_all:
            
        # filter
        wt_U = ds['U'].isel(depth=depth).window
        wt_U.set(n={'lat': grid_num, 'lon_g': grid_num}, window={'lat':'hanning', 'lon_g':'hanning'}, chunks = {'lat':1, 'lon_g':177,'time': 9415})
        ds_U_filter = wt_U.convolve()

        wt_V = ds['V'].isel(depth=depth).window
        wt_V.set(n={'lat_g': grid_num, 'lon': grid_num}, window={'lat_g':'hanning', 'lon':'hanning'}, chunks = {'lat_g':1, 'lon':177,'time': 9415})
        ds_V_filter = wt_V.convolve()

        wt_U2 = (ds['U'].isel(depth=depth)**2).window
        wt_U2.set(n={'lat': grid_num, 'lon_g': grid_num}, window={'lat':'hanning', 'lon_g':'hanning'}, chunks = {'lat':1, 'lon_g':177,'time': 9415})
        ds_U2_filter = wt_U2.convolve()

        wt_V2 = (ds['V'].isel(depth=depth)**2).window
        wt_V2.set(n={'lat_g': grid_num, 'lon': grid_num}, window={'lat_g':'hanning', 'lon':'hanning'}, chunks = {'lat_g':1, 'lon':177,'time': 9415})
        ds_V2_filter = wt_V2.convolve()

        uv = gridx.interp(ds.V.isel(depth=depth),'Y')*gridx.interp(ds.U.isel(depth=depth),'X')
        wt_UV = uv.window
        wt_UV.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat':1, 'lon':177,'time': 9415})
        ds_UV_filter = wt_UV.convolve()

        # Pi_omega
        du_dx = gridx.diff(ds_U_filter, 'X')/gridx.interp(ds.dxG, 'Y')
        dv_dy = gridx.diff(ds_V_filter, 'Y')/gridx.interp(ds.dyG, 'X')
        du_dy = gridx.interp(gridx.interp(gridx.diff(ds_U_filter, 'Y')/gridx.interp(ds.dyC, 'X'), 'X'), 'Y')
        dv_dx = gridx.interp(gridx.interp(gridx.diff(ds_V_filter, 'X')/gridx.interp(ds.dxC, 'Y'), 'X'), 'Y')

        PI_1 = (gridx.interp(ds_U2_filter, 'X')-gridx.interp(ds_U_filter**2, 'X'))*du_dx  # [f(u^2) - f(u)^2]*f(du/dx)
        PI_2 = (gridx.interp(ds_V2_filter, 'Y')-gridx.interp(ds_V_filter**2, 'Y'))*dv_dy  # [f(v^2) - f(v)^2]*f(dv/dy)
        PI_3 = (ds_UV_filter - gridx.interp(ds_U_filter,'X')*gridx.interp(ds_V_filter,'Y') )*(du_dy+dv_dx) # [f(uv) - f(u)f(v)]*[f(du/dy) + f(dv/dx)]

        PI_cg_single = -(PI_1+PI_2+PI_3)
        chunks = {'lat': 1, 'lon': 177, 'time':9415}

        # save
        PI_cg_single = PI_cg_single.chunk(chunks)
        PI_cg_single = np.real(PI_cg_single).rename('PI_cg_single')     
        PI_cg_single = PI_cg_single.to_dataset()
        file_out = out_dir_store+'/%s_d%02d.zarr'%(vv,depth)
        PI_cg_single.to_zarr(file_out, mode='w')                    
        print('%s, grid_num=%d depth=%d - allready processed'%(vv,grid_num,depth))
```

# 3. Calculate $\Pi_\omega$ using the Coarse-graining method (using a temporal filter)
## KE flux $\Pi_\omega$ at 1 day, 5 days and 30 days

```python
out_dir_store = '/home1/scratch/xyu/KE/Pi_omega/'
VV = ['PI_cg_1d','PI_cg_5d','PI_cg_30d']
Time_length = [24*1, 24*5, 24*30]
depth_all = range(52)
```

```python
# 23 mins for 1d
# 23 mins for 5d 
# 80 mins for 30d


for vv,time_length in zip(VV,Time_length):
    
    for depth in depth_all:
        
        # filter
        wt_U = ds['U'].isel(depth=depth).window
        wt_U.set(n={'time': time_length}, window={'time':'hanning'},chunks = {'lat':1, 'lon_g':177,'time': 9415})
        ds_U_filter = wt_U.convolve()

        wt_V = ds['V'].isel(depth=depth).window
        wt_V.set(n={'time': time_length}, window={'time':'hanning'},chunks = {'lat_g':1, 'lon':177,'time': 9415})
        ds_V_filter = wt_V.convolve()

        wt_U2 = (ds['U'].isel(depth=depth)**2).window
        wt_U2.set(n={'time': time_length}, window={'time':'hanning'},chunks = {'lat':1, 'lon_g':177,'time': 9415})
        ds_U2_filter = wt_U2.convolve()

        wt_V2 = (ds['V'].isel(depth=depth)**2).window
        wt_V2.set(n={'time': time_length}, window={'time':'hanning'},chunks = {'lat_g':1, 'lon':177,'time': 9415})
        ds_V2_filter = wt_V2.convolve()

        uv = gridx.interp(ds.V.isel(depth=depth),'Y')*gridx.interp(ds.U.isel(depth=depth),'X')
        wt_UV = uv.window
        wt_UV.set(n={'time': time_length}, window={'time':'hanning'},chunks = {'lat':1, 'lon':177,'time': 9415})
        ds_UV_filter = wt_UV.convolve()

        # Pi_omega
        du_dx = gridx.diff(ds_U_filter, 'X')/gridx.interp(ds.dxG, 'Y')
        dv_dy = gridx.diff(ds_V_filter, 'Y')/gridx.interp(ds.dyG, 'X')
        du_dy = gridx.interp(gridx.interp(gridx.diff(ds_U_filter, 'Y')/gridx.interp(ds.dyC, 'X'), 'X'), 'Y')
        dv_dx = gridx.interp(gridx.interp(gridx.diff(ds_V_filter, 'X')/gridx.interp(ds.dxC, 'Y'), 'X'), 'Y')

        PI_1 = (gridx.interp(ds_U2_filter, 'X')-gridx.interp(ds_U_filter**2, 'X'))*du_dx  # [f(u^2) - f(u)^2]*f(du/dx)
        PI_2 = (gridx.interp(ds_V2_filter, 'Y')-gridx.interp(ds_V_filter**2, 'Y'))*dv_dy  # [f(v^2) - f(v)^2]*f(dv/dy)
        PI_3 = (ds_UV_filter - gridx.interp(ds_U_filter,'X')*gridx.interp(ds_V_filter,'Y') )*(du_dy+dv_dx) # [f(uv) - f(u)f(v)]*[f(du/dy) + f(dv/dx)]

        PI_cg_single = -(PI_1+PI_2+PI_3)
        chunks = {'lat': 1, 'lon': 177, 'time':9415}
        PI_cg_single = PI_cg_single.chunk(chunks)
        PI_cg_single = np.real(PI_cg_single).rename('PI_cg_single')   
        # save
        PI_cg_single = PI_cg_single.to_dataset()
        file_out = out_dir_store+'/%s_d%02d.zarr'%(vv,depth)
        PI_cg_single.to_zarr(file_out, mode='w')                    
        print('%s, time_length=%d depth=%d - allready processed'%(vv,time_length,depth))
```

```python
def filter_2D(v, grid_num):
        
    wt = v.window    
    if "U" in v:
        wt.set(n={'lat': grid_num, 'lon_g': grid_num}, window={'lat':'hanning', 'lon_g':'hanning'}, chunks = {'lat':1, 'lon_g':177,'time': 9415, 'depth':1})
    elif "V" in v:
        wt.set(n={'lat_g': grid_num, 'lon': grid_num}, window={'lat_g':'hanning', 'lon':'hanning'}, chunks = {'lat_g':1, 'lon':177,'time': 9415, 'depth':1})
    elif "UV" in v:
        wt.set(n={'lat': grid_num, 'lon': grid_num}, window={'lat':'hanning', 'lon':'hanning'}, chunks = {'lat':1, 'lon':177,'time': 9415, 'depth':1})
    
    v_filter = wt.convolve()
    
    return v_filter   
```
